## Functions

```python
def loadGloveModel(file):
    f = open(file,'r')
    model = dict()
    
    for line in f:
        splitLines = line.split()
        word = splitLines[0]
        embedding = np.array([float(value) for value in splitLines[1:]])
        model[word] = embedding
    
    return model
```

```python
def remove_number(text):
    no_number = re.sub(r'\d+', ' ', text)
    return no_number

def remove_username(text):
    no_username = re.sub(r'\@[^\s]+', ' ', text)
    return no_username

def remove_stopwords(text):
    no_stopwords = [word for word in text.split() if word not in sw]
    return ' '.join(no_stopwords)

def remove_urls(text):
    no_urls = re.sub(r'http[^\s]+|www\.[^\s]+', ' ', text)
    return no_urls

def remove_accent(text):
    no_accent = unidecode(text)
    return no_accent

def remove_punctuation(text):
    no_punctuation = re.sub(r'[^\w\s]', ' ', text)
    return no_punctuation

def remove_html(text):
    no_html = re.sub(r'<.*?>', ' ', text)
    return no_html

def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    no_emoji = emoji_pattern.sub(r'', text)
    return no_emoji
    
def format_text(text):
    no_multiple_spaces = re.sub(r'\s+', ' ', text)
    no_multiple_spaces = no_multiple_spaces.strip()
    return no_multiple_spaces

def remove_one_length_word(text):
    no_one_length_word = re.sub(r'\b[a-z]\b', ' ', text)
    return no_one_length_word

def format_slangs(text):
    text = text.replace("i've", 'i have')
    text = text.replace("i'm", 'i am')
    text = text.replace("what's", 'what is')
    text = text.replace("won't", 'will not')
    text = text.replace("hwy", 'highway')
    text = text.replace("haven't", 'have not')
    text = text.replace("you're", 'you are')
    text = text.replace("im", 'i am')
    text = text.replace("don't", 'do not')
    text = text.replace("doesn't", 'does not')
    text = text.replace("he'll", 'he will')
    text = text.replace("he's", 'he is')
    text = text.replace("she's", 'she is')
    text = text.replace("didn't", 'did not')
    text = text.replace("ain't", 'am not')
    text = text.replace("i'd", 'i would')
    text = text.replace("we're", 'we are')
    text = text.replace("it's", 'it is')
    text = text.replace("hasn't", 'has not')
    text = text.replace("bc", 'because')
    text = re.sub(r'\bu\b', '\byou\b', text)
    text = re.sub(r'\by\b', '\bwhy\b', text)
    return text

def remove_laugh(text):
    no_laugh = re.sub(r'\b(a|e|i)?([ha|he|hi])+(h)?(a|e|i)?\b', ' ', text)
    return no_laugh

def standardization(text):
    no_duplicate_letters= re.sub(r'([a-z])\1+', r'\1', text)
    return no_duplicate_letters

def preprocessing(data):
    data = pd.Series(data)
    data = data.apply(str)
    data = data.apply(lambda x: x.lower())
    data = data.apply(remove_html)
    data = data.apply(remove_emoji)
    data = data.apply(remove_urls)
    data = data.apply(remove_username)
    data = data.apply(remove_number)
    data = data.apply(remove_stopwords)
    data = data.apply(format_slangs)
    data = data.apply(remove_punctuation)
    data = data.apply(remove_accent)
    data = data.apply(remove_one_length_word)
    data = data.apply(standardization)
    data = data.apply(remove_laugh)
    data = data.apply(format_text)
    return data
```

```python
def words_per_text(text):
    text = str(text)
    return len(text.split())
```

## Normal Embedding

```python
train['cleaned_text'] = preprocessing(train['text'])
train = train.dropna()
train = train.drop_duplicates()
train = train.reset_index()
train = train.drop(columns=['index'])
train.head()
```

```python
train.shape
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train['cleaned_text'])
```

```python
vocab_size = len(tokenizer.word_index) + 1
print(vocab_size)
```

```python
tokenizer.word_index
```

```python
train['tokenized_text'] = tokenizer.texts_to_sequences(train['cleaned_text'])
train['tokenized_text'].head()
```

```python
X = train['tokenized_text']
y = train['target']
```

### Checking mean words per text after preprocessing and before

```python
temp = pd.DataFrame()
temp['depois'] = train['cleaned_text'].apply(words_per_text)
temp['antes'] = train['text'].apply(words_per_text)
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(data=temp, x="antes", color="skyblue", label="Antes", bins=15, kde=True)
sns.histplot(data=temp, x="depois", color="red", label="Depois", bins=15, kde=True)
plt.legend() 
plt.show()
```

```python
print("Antes: {}, Depois: {}".format(temp['antes'].mean(), temp['depois'].mean()))
del temp
```

```python
X = pad_sequences(X, maxlen=10, padding='post')
```

```python
X.shape
```

### Training the model

```python
accuracies = []
losses = []
k = list(range(1, 6))

for train, test in kf.split(X, y):
    
    print('\n-----------------------\n')
    
    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, 
                        output_dim=100, 
                        input_length=10))
    model.add(Conv1D(32, 3, activation='relu', padding='same'))
    model.add(MaxPooling1D())
    model.add(Conv1D(64, 5, activation='relu', padding='same'))
    model.add(MaxPooling1D())
    model.add(Dropout(rate=0.2))
    model.add(Flatten())
    model.add(Dense(64, activation='sigmoid'))
    model.add(Dense(128, activation='sigmoid'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    model.fit(X[train], y[train], epochs=5, batch_size=32)
    score = model.evaluate(X[test], y[test])
    accuracies.append(score[1])
    losses.append(score[0])
```

```python
results = pd.DataFrame({
    'K': k,
    'Accuracy': accuracies,
    'Loss': losses
})

plt.figure(figsize=(8, 6))
sns.lineplot(data=results, x='K', y='Accuracy', label='accuracy')
sns.lineplot(data=results, x='K', y='Loss', label='loss')
plt.legend()
plt.show()
```

## Glove Embedding

```python
%%time
glove = loadGloveModel('glove.6B.300d.txt')
```

```python
embedding = np.zeros((vocab_size, 300))

for word, index in tokenizer.word_index.items():
    try:
        embedding[index] = glove[word]
    except:
        continue
```

### Training the model

```python
accuracies = []
losses = []
k = list(range(1, 6))
fold = 1

for train, test in kf.split(X, y):
    
    print('\n -------------------- Fold: {} --------------------  \n'.format(fold))
    
    model_glove = Sequential()
    model_glove.add(Embedding(input_dim=vocab_size, 
                        output_dim=300, 
                        input_length=10,
                        weights=[embedding],
                        trainable=False))
    model_glove.add(Conv1D(32, 3, activation='relu', padding='same'))
    model_glove.add(MaxPooling1D())
    model_glove.add(Conv1D(64, 5, activation='relu', padding='same'))
    model_glove.add(MaxPooling1D())
    model_glove.add(Dropout(rate=0.2))
    model_glove.add(Flatten())
    model_glove.add(Dense(64, activation='sigmoid'))
    model_glove.add(Dense(128, activation='sigmoid'))
    model_glove.add(Dense(1, activation='sigmoid'))
    model_glove.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    model_glove.fit(X[train], y[train], epochs=5, batch_size=32)
    score = model_glove.evaluate(X[test], y[test])
    accuracies.append(score[1])
    losses.append(score[0])
    fold += 1
```

```python
results = pd.DataFrame({
    'K': k,
    'Accuracy': accuracies,
    'Loss': losses
})

plt.figure(figsize=(8, 6))
sns.lineplot(data=results, x='K', y='Accuracy', label='accuracy')
sns.lineplot(data=results, x='K', y='Loss', label='loss')
plt.legend()
plt.show()
```

### Prediction

```python
prediction = model_glove.predict_classes(test_df, batch_size=32)
predicted = [p[0] for p in prediction]
```

### Saving the results

```python
results = pd.DataFrame()
results['id'] = id_test
results['target'] = predicted
results.to_csv('submission_cnn_glove.csv', index=False)
```
