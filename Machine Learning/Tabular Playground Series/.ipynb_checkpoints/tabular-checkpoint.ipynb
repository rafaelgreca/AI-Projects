## Models

```python
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import ElasticNet

models = []
accuracies = []

rf_model = RandomForestRegressor(random_state=19, n_jobs=-1)
xgb_model = xgb.XGBRegressor(random_state=19, n_jobs=-1)
linreg_model = LinearRegression(n_jobs=-1)
lasso_model = Lasso(random_state=19)
extratree_model = ExtraTreesRegressor(random_state=19, n_jobs=-1)
decisiontree_model = DecisionTreeRegressor(random_state=19)
elastic_net = ElasticNet(random_state=19)
```

```python
X = train_data.drop(columns=['target'])
y = train_data['target']
```

### Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=19)

rf_model.fit(X_train, y_train)
predicted = rf_model.predict(X_test)
accuracies.append(np.sqrt(mean_squared_error(y_test, predicted)))
models.append("Random Forest")
```

### XGBoost

```python
xgb_model.fit(X_train, y_train)
predicted = xgb_model.predict(X_test)
accuracies.append(np.sqrt(mean_squared_error(y_test, predicted)))
models.append("XGBoost")
```

### Linear Regression

```python
linreg_model.fit(X_train, y_train)
predicted = linreg_model.predict(X_test)
accuracies.append(np.sqrt(mean_squared_error(y_test, predicted)))
models.append("Linear Regression")
```

### Lasso

```python
lasso_model.fit(X_train, y_train)
predicted = lasso_model.predict(X_test)
accuracies.append(np.sqrt(mean_squared_error(y_test, predicted)))
models.append("Lasso")
```

### Extra Trees

```python
extratree_model.fit(X_train, y_train)
predicted = extratree_model.predict(X_test)
accuracies.append(np.sqrt(mean_squared_error(y_test, predicted)))
models.append("Extra Tree")
```

### Decision Tree

```python
decisiontree_model.fit(X_train, y_train)
predicted = decisiontree_model.predict(X_test)
accuracies.append(np.sqrt(mean_squared_error(y_test, predicted)))
models.append("Decision Tree")
```

### Elastic Net

```python
elastic_net.fit(X_train, y_train)
predicted = elastic_net.predict(X_test)
accuracies.append(np.sqrt(mean_squared_error(y_test, predicted)))
models.append("Elastic Net")
```

## Pivot Table

```python
df = pd.DataFrame({
    "Model": models,
    "Accuracy": accuracies
})

pd.pivot_table(df, index="Model")
```

## Submission

```python
submission = pd.DataFrame({
    'id': test_data['id'],
    'target': xgb_model.predict(test_data.drop(columns=['id']))
})

submission.to_csv('submission_tabular.csv', index=False)
```

## Tunning XGBoost parameters using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [200],
    'min_child_weight': [1, 5, 10],
    'gamma': [0.5, 1, 1.5, 2, 5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'max_depth': [3, 4, 5]
}

grid = GridSearchCV(param_grid=params,
                    estimator=xgb_model, 
                    n_jobs=-1, 
                    cv=3,
                    verbose=3)

grid.fit(X_train, y_train)
predicted = grid.predict(X_test)
accuracy = np.sqrt(mean_squared_error(y_test, predicted))
accuracy
```
