## Checking for Outliers

```python
numerical_columns = train_data.select_dtypes(exclude="object")
numerical_columns.drop(columns=["SalePrice"], inplace=True)
numerical_columns.shape
```

```python
figure = plt.figure(figsize=(10, 20))

for i in range(len(numerical_columns.columns)):
    figure.add_subplot(12, 3, i+1)
    sns.boxplot(numerical_columns.iloc[:, i])

plt.tight_layout()
plt.show()

'''
possible outliers detected in columns:
LotFrontage (>250)
LotArea (>100000)
BsmtFinSF1 (>3000)
TotalBsmtSF (>4000)
1stFlrSF (>4000)
GrLivArea (>4000)
MiscVal (>6000)
'''
```

```python
figure = plt.figure(figsize=(10, 20))

for i in range(len(numerical_columns.columns)):
    figure.add_subplot(12, 3, i+1)
    sns.scatterplot(numerical_columns.iloc[:, i], train_data['SalePrice'])

plt.tight_layout()
plt.show()

'''
outliers detected in columns:
LotFrontage (>250)
LotArea (>100000)
BsmtFinSF1 (>3000)
TotalBsmtSF (>4000)
1stFlrSF (>4000)
GrLivArea (>4000) and SalePrice (>500000)
LowQualFinSF (>550)
'''
```

```python
#removing the outliers

train_data = train_data[train_data['LotFrontage'] < 250]
train_data = train_data[train_data['LotArea'] < 100000]
train_data = train_data[train_data['BsmtFinSF1'] < 3000]
train_data = train_data[train_data['TotalBsmtSF'] < 4000]
train_data = train_data[train_data['1stFlrSF'] < 4000]
train_data = train_data[~((train_data['GrLivArea'] > 4000) & (train_data['SalePrice'] < 500000))]
train_data = train_data[train_data['LowQualFinSF'] < 550]
```

```python
train_data.shape
```

## Checking Correlations

```python
plt.figure(figsize=(20, 15))
corr = train_data.corr()
sns.heatmap(corr[['SalePrice']], annot=True)

columns_to_drop = ['MSSubClass', 'OverallCond', 'BsmtFinSF2', 'LowQualFinSF', 'BsmtHalfBath', 'YrSold',
                   'MoSold', 'MiscVal', 'PoolArea', 'ScreenPorch', '3SsnPorch', 'EnclosedPorch', 'KitchenAbvGr']
```

```python
train_data.drop(columns=columns_to_drop, inplace=True)
test_data.drop(columns=columns_to_drop, inplace=True)
```

```python
plt.figure(figsize=(20, 15))
corr = train_data.corr()
sns.heatmap(corr, annot=True)

'''
GarageYrBuilt has a high correlation with YearBuilt. We will keep YearBuild because it has a
higher correlation with SalePrice.

1stFlrSF has a high correlation with TotalBsmtSF. We will keep TotalBsmtSF because it has a
higher correlation with SalePrice.

GrLivArea has a high correlation with TotRmsAbvGrd. We will keep GrLivArea because it has a
higher correlation with SalePrice.

GarageCars has a high correlation with GarageArea. We will keep GarageCars because it has a
higher correlation with SalePrice.
'''
```

```python
train_data.drop(columns=['TotRmsAbvGrd', 'GarageArea', '1stFlrSF', 'GarageYrBlt'], inplace=True)
test_data.drop(columns=['TotRmsAbvGrd', 'GarageArea', '1stFlrSF', 'GarageYrBlt'], inplace=True)
```

## Models (No Feature Selection)

```python
'''
models we are gonna test:
random forest, xgboost, gradient descent, svm, linear regression, logistic regression, decision tree
'''
import xgboost as xgb
from sklearn import svm
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

accuracies = []
models = []

random_forest = RandomForestRegressor(random_state=19)
svm_model = svm.SVR()
xgb_model = xgb.XGBRegressor(random_state=19)
gradient = SGDRegressor(random_state=19, early_stopping=True)
linear_regression = LinearRegression()
decision_tree = DecisionTreeRegressor(random_state=19)
```

### Random Forest

```python
random_forest.fit(X_train, y_train)
y_predicted = random_forest.predict(X_test)
accuracies.append(mean_absolute_error(y_test, y_predicted))
models.append("Random Forest")
```

### XGBoost

```python
xgb_model.fit(X_train, y_train)
y_predicted = xgb_model.predict(X_test)
accuracies.append(mean_absolute_error(y_test, y_predicted))
models.append("XGBoost")
```

### Gradient Descent

```python
gradient.fit(X_train, y_train)
y_predicted = gradient.predict(X_test)
accuracies.append(mean_absolute_error(y_test, y_predicted))
models.append("Gradient Descent")
```

### SVM

```python
svm_model.fit(X_train, y_train)
y_predicted = svm_model.predict(X_test)
accuracies.append(mean_absolute_error(y_test, y_predicted))
models.append("SVM")
```

### Linear Regression

```python
linear_regression.fit(X_train, y_train)
y_predicted = linear_regression.predict(X_test)
accuracies.append(mean_absolute_error(y_test, y_predicted))
models.append("Linear Regression")
```

### Logistic Regression

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_predicted = logistic.predict(X_test)
accuracies.append(mean_absolute_error(y_test, y_predicted))
models.append("Logistic Regression")
```

## Decision Tree

```python
decision_tree.fit(X_train, y_train)
y_predicted = decision_tree.predict(X_test)
accuracies.append(mean_absolute_error(y_test, y_predicted))
models.append("Decision Tree")
```

### Stacking Regressor

```python
from sklearn.ensemble import StackingRegressor

estimators = [
    ('rf', random_forest),
    ('xgb', xgb_model),
    ('sgb', gradient),
    ('svm', svm_model),
    ('linear_regression', linear_regression)
]

stacking = StackingRegressor(estimators = estimators, n_jobs=-1)
stacking.fit(X_train, y_train)
y_predicted = stacking.predict(X_test)
accuracies.append(mean_absolute_error(y_test, y_predicted))
models.append("Stacking Regressor")
```
